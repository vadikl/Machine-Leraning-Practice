```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np
#sns.set_style("darkgrid") # can set seaborn style

%matplotlib inline
%config InlineBackend.figure_format = 'svg' # Can control quality of figures png - poor quality, svg - high quality
```

### Main Plot Elements


### Basic plot for numerical values

```python
x = np.linspace(0, 10, 50)
y1 = x
y2 = [i**2 for i in x]

plt.figure(figsize=(12,5)) # Firstly,we need to define a figure (box) where we will be plotting

plt.plot(x, y1, 'r--' , x, y2, 'b')
plt.title('Basic Plot', fontsize=14)
plt.xlabel('x', fontsize=14, color='b')
plt.ylabel('y', fontsize=14, color='b')
plt.grid()
#fig.savefig("filename.png", dpi=200) for saving the results 
```

### Basic barplot for Categorical Values

```python
fruits = ["apple", "peach", "orange", "bannana", "melon"]
counts = [34, 25, 43, 31, 17]
plt.bar(fruits, counts)
plt.title("Fruits!")
plt.xlabel("Fruit")
plt.ylabel("Count")
```

### Subplot

```python
x = [1, 5, 10, 15, 20]
y = [1, 7, 3, 5, 11]

fig, ax = plt.subplots(1,2,figsize=(12,5)) # return a tuple with axes and a figure 

fig.suptitle("Two different point styles")
ax[0].plot(x,y,'ro',label='points')
ax[0].set_xlabel('x1')
ax[0].set_ylabel('y1')
ax[0].legend(loc='best')
ax[1].plot(x,y,'bx',label='crosses')
ax[1].set_xlabel('x2')
ax[1].set_ylabel('y2')
ax[1].legend(loc='center');
```

### GridSpec

```python
import matplotlib.gridspec as gridspec
```

```python
x = [1, 2, 3, 4, 5]
y1 = [9, 4, 2, 4, 9]
y2 = [1, 7, 6, 3, 5]

fig = plt.figure(figsize=(7, 3))
gs = gridspec.GridSpec(ncols=2, nrows=1, figure=fig)

fig_ax_1 = fig.add_subplot(gs[0, 0])
plt.plot(x, y1)
fig_ax_2 = fig.add_subplot(gs[0, 1])
plt.plot(x, y2);
```

```python
x = [1, 2, 3, 4, 5]
y1 = [9, 4, 2, 4, 9]
y2 = [1, 7, 6, 3, 5]
y3 = [-7, -4, 2, -4, -7]
```

```python
fig = plt.figure(figsize=(9, 4))
gs = fig.add_gridspec(2, 2)

fig_ax_1 = fig.add_subplot(gs[0, :])
plt.plot(x, y2)
fig_ax_2 = fig.add_subplot(gs[1, 0])
plt.plot(x, y1)
fig_ax_3 = fig.add_subplot(gs[1, 1])
plt.plot(x, y3);
```
