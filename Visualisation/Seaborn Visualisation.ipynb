```python
data = pd.read_csv('C:/Users/vladislav/Desktop/ODS/Telephones.csv')
data.head()
```

### Distribution Plot

```python
sns.distplot(data['total eve minutes']);
```

### Scatter Plot 

```python
sns.scatterplot(data['total day minutes'],data['total eve minutes'],hue=data['churn']);
```

### Regression Plot

```python
sns.regplot(data['total intl calls'],data['total intl minutes']);
```

```python
sns.lmplot(x='total day minutes',y='total eve minutes',hue='churn',data=data);
```

### Swarm plot (Categorical Scatter Plot)

```python
sns.swarmplot(data['churn'],data['total day minutes']);
```

### Bar Plot

```python
sns.barplot(data['international plan'],data['churn']);
```

### Count Plot

```python
sns.countplot(data['churn']);
```

### Kde Plot

```python
sns.kdeplot(data['total intl minutes'],shade=True);
```

```python
sns.kdeplot(data['total intl minutes'],shade=True);
sns.kdeplot(data['total intl charge'],shade=True);
sns.kdeplot(data['total intl calls'],shade=True);
```

### Box Plot

```python
sns.boxplot(data['churn'],data['total intl minutes']);
```

### Violin Plot

```python
sns.violinplot(data['churn'],data['total intl minutes']);
```
